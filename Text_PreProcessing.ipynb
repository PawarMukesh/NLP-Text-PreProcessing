## Remove Default StopWords
Stop words is that word that do not contribut in meaning of sentence

```python
# Importing nltk library to remove the stop words
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Downloading the stopword from nltk library
nltk.download('stopwords')
nltk.download('punkt')

# Write a function to remove the stopword
def remove_stopwords(text):
  stop_words = set(stopwords.words('english'))
  word_token = word_tokenize(text)
  filter_text = [word for word in word_token if word not in stop_words]
  return filter_text

ex_text = "Data is the new oil. A.I is the last invention"
remove_stopwords(ex_text)
```

## Stemming
Stemming is the process of getting the root of word, we will remove the prefix and suffix of word, like: Mangoes----> Mango

```python
# Without tokenization you can not perform stemming
# importing nltk porter stemmer

from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize

stem1 = PorterStemmer()

# Stem words in the list of tokenize word
def s_word(text):
  word_tokens = word_tokenize(text)
  stems = [stem1.stem(word) for word in word_tokens]
  return stems

text = 'Data is the new revolution in the World, in a day one individual would generate terabytes of data.'
print(s_word(text))

```

## Lemmatization

As stemming, lemmatization do the same but the only difference is that lemmatization ensures that root word belongs to the language. Because of the use of lemmatization we will get the valid words. In NLTK(Natural language Toolkit), we use WordLemmatizer to get the lemmas of words. We also need to provide a context for the lemmatization.So, we added pos(parts-of-speech) as a parameter. 

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('omw-1.4')
lemma = WordNetLemmatizer()
nltk.download('wordnet')  # Contain all roots word

# lemmatiza string
def lemmatize_word(text):
  word_token = word_tokenize(text)
  lemmas = [lemma.lemmatize(word, pos='v') for word in word_token] # Provide context i.e part of speech(pos)
  return lemmas

```

```python
text = 'Data is the new revolution in the World, in a day one individual would generate terabytes of data.'
lemmatize_word(text)
```

## Parts of speech (POS) Tagging

The pos(parts of speech) explain you how a word is used in a sentence. In the sentence, a word have different contexts and semantic meanings. The basic natural language processing(NLP) models like bag-of-words(bow) fails to identify these relation between the words. For that we use pos tagging to mark a word to its pos tag based on its context in the data. Pos is also used to extract relationship between the words. 

```python

# importing tokenize library
from nltk.tokenize import word_tokenize 
from nltk import pos_tag 
nltk.download('averaged_perceptron_tagger')
  
# convert text into word_tokens with their tags 
def pos_tagg(text): 
    word_tokens = word_tokenize(text) 
    return pos_tag(word_tokens) 
  
pos_tagg('Are you afraid of something?') 

# IN: preposition
# NN:  noun sigular
```

In the above example NNP stands for Proper noun, PRP stands for personal noun, IN as Preposition. We can get all the details pos tags using the Penn Treebank tagset.

```python
import nltk
# downloading the tagset  
nltk.download('tagsets') 
  
# extract information about the tag 
nltk.help.upenn_tagset('PRP') # You Can change the tag and ge the tag information
```

## Named entity recognition
It is used to extract the information from unstrucutre text it is used to classy the entities, which is present in the text into categories like person, orgnization, event, places etc.. this will give you a detail knowldge about text and relationship between different entities

```python
# Importing tokenization and chunk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def name_entity(text):

  #Tokenize the text
  word_tokens=word_tokenize(text)

  # Pos tagging of words
  pos = pos_tag(word_tokens)

  # Named entity
  print(ne_chunk(pos))

text = 'Indian win the 2007 word cup againest the pakistan in south africa.'
name_entity(text) 
```

## Chunking
Chunking is the process of extracting phrases from unstructure text and give them more structure to it, we also called them shallow parsing.

```python
# Importing the library
from nltk.tokenize import word_tokenize
from nltk import pos_tag

## Here we are defining rhe function with text and regular expression represnting grammer as a parameter
def chunk(text, grammer):
  word_token = word_tokenize(text)
  pos = pos_tag(word_token)

  # Create a chunk parser using grammer
  chunkparser = nltk.RegexpParser(grammer)

  # Test it on the list of words and token with tag pos
  tree = chunkparser.parse(pos)

  for subtree in tree.subtrees():
    print(subtree)
    #tree.draw()

sentence = 'the little red parrot is flying in the sky'
grammer = 'NP: {<DT>?<JJ>*<NN>}'
chunk(sentence,grammer)
```


In the above example, we defined the grammer by using the regular expression rule, this rule tells you that NP(Noun Phrase) chunkshould be formed whenever the chunker find the optional determiner (DJ) followd by any no of adjective and then a NN(Noun)

# Word Count
I am assuming you have the understanding of tokenization,the first figure we can calculate is the word frequency.By *word frequency* we can find out how many times each tokens appear in the text. When talking about word frequency, we distinguished between *types* and *tokens*.Types are the distinct words in a corpus, whereas tokens are the words, including repeats. Let's see how this works in practice.

Let's take an example for better understanding:

“There is no need to panic. We need to work together, take small yet important measures to ensure self-protection,” the Prime Minister tweeted.

How many tokens and types are there in above sentences?

Let's use Python for calculating these figures. First, tokenize the sentence by using the tokenizer which uses the non-alphabetic characters as a separator.

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize.regexp import WhitespaceTokenizer

m = "'There is no need to panic. We need to work together, take small yet important measures to ensure self-protection,' the Prime Minister tweeted."
```

```python
# Getting the total no of token are present in sentense

token = WhitespaceTokenizer().tokenize(m)
print(token)
```

```python
# Getting the length of token
len(token)
```

```python
# Remove the repeted word 
remove = set(token)
print(remove)
```

```python
# Then getting length of token after removing the repeted word
len(remove)
```

```python
# Now we are perform same operation with the help of different tokenizer
from nltk.tokenize.regexp import WordPunctTokenizer  # This is count punctuation as a token

my_st = "'There is no need to panic. We need to work together, take small yet important measures to ensure self-protection,' the Prime Minister tweeted."

tokens = WordPunctTokenizer().tokenize(my_st)
print(tokens)


# Getting the length of tokens
print("Length of Tokens:", len(tokens))

# Remove the repeted word 
r = set(tokens)
print("After Removing repeted word", r)

# Getting the length after removing the repeted word
print("Length of tokens after removing the repeted words",len(r))   # But the problem assosiate with wordPunctTokenizer is they can count punctuation as a token
```
